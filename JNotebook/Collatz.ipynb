# Indice de Collatz

## Codigo realizado en Python

```python
import matplotlib.pyplot as plt
import numpy as np

def collatz(n):
    lista = []
    while n != 1:
        lista.append(n)
        if n % 2 == 0:
            n = n / 2
        else:
            n = 3 * n + 1
    lista.append(1)
    return lista

def graficar():
    x = np.linspace(1, 1000000, 1000000)
    y = []
    for i in range(1, 1000001):
        y.append(len(collatz(i)))
    plt.plot(x, y, 'b-')
    plt.title('Grafica de Collatz')
    plt.xlabel('Numero')
    plt.ylabel('Indice de Collatz')
    plt.show()

if __name__ == '__main__':
    graficar()
```
